```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random 
```

```python
left=np.arange(1,11)
right=np.array([random.randint(3,8) for r in range(10)])
```

```python
print(left)
print(right)
```

```python
test = np.vstack((left,right)).T
np.dtype([('test', np.float)])
np.savetxt('test.csv', test, delimiter = ',')  
```

```python
print(test)
```

```python
print(np.shape(test))
```

```python
plt.figure(1,facecolor = 'red',figsize=(10,8))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Numpy Graph')
plt.plot(test[:,0],test[:,1])
```

```python
plt.show()
```
